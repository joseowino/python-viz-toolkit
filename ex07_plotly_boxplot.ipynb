7: Plotly Box plots

The goal of this exercise is to learn to use Plotly to plot Box Plots. It is a method for graphically depicting groups of numerical data through their quartiles and values as min, max. It allows comparing quickly some variables.

Generate 3 random arrays from a normal distribution. Keep one as it is and for the other two arrays add respectively 1 and 2 to the normal distribution.

y0 = np.random.randn(50)
y1 = np.random.randn(50) + 1 # shift mean
y2 = np.random.randn(50) + 2
In the same figure, plot the 3 distributions as box plots as shown in the image. In this exercise the style is not important.

The plot has to contain:

The title

The legend

```python
# Configure

import numpy as np
import pandas as pd
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode

init_notebook_mode(connected=True)

print("Configured Successfully")
```

```python
# Data

y0 = np.random.randn(50)
y1 = np.random.randn(50) + 1
y2 = np.random.randn(50) + 2

print(pd.DataFrame(
    {
        "y0": y0,
        "y1": y1,
        "y2": y2
    }
))
```

```python
# Chart

fig = go.Figure()

fig.add_trace(go.Box(y=y0, name='y0'))
fig.add_trace(go.Box(y=y1, name='y1'))
fig.add_trace(go.Box(y=y2, name='y2'))

fig.update_layout(
    title='Box plot of 3 distributions',
    showlegend=True
)

fig.show()
```
