The plot has to contain:

the title

```python
# Import Packages
import pandas as pd
import matplotlib.pyplot as plt

print("All Packages imported successfuly...")
```

```python
df = pd.DataFrame({
    'name':['christopher','marion','maria','mia','clement','randy','remi'],
    'age':[70,30,22,19,45,33,20],
    'gender':['M','F','F','F','M','M','M'],
    'state':['california','dc','california','dc','california','new york','porto'],
    'num_children':[4,2,1,0,3,1,0],
    'num_pets':[5,1,0,2,2,2,3]
})

print(df)
```

```python
df.plot(
    x='age',
    y='num_children',
    kind='scatter',
    title='Age and number of children',
    xlabel='age',
    ylabel='num_children',
    figsize=(8, 5)
)

plt.tight_layout()
plt.show()
```
