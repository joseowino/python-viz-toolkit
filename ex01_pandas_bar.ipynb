```python
# Packages
import pandas as pd
import matplotlib.pyplot as plt

print("Packages imported successfully...\n")
```

```python
df = pd.DataFrame({
    'name':['christopher','marion','maria','mia','clement','randy','remi'],
    'age':[70,30,22,19,45,33,20],
    'gender':['M','F','F','F','M','M','M'],
    'state':['california','dc','california','dc','california','new york','porto'],
    'num_children':[2,0,0,3,8,1,4],
    'num_pets':[5,1,0,5,2,2,3]
})

print(df)
```

```python
df.plot(
    kind='bar', 
    x='name', 
    y=['num_children', 'num_pets'], 
    title='Number of children and pets per person',
    figsize=(10, 5)
)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```
