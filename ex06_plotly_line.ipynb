```python
# Packages

import numpy as np
import pandas as pd
import plotly.express as px

print("Packages imported Successfully")
```

```python
# Data

returns = np.random.randn(50)
price = 100 + np.cumsum(returns)

dates = pd.date_range(start='2020-09-01', periods=50, freq='B')
df = pd.DataFrame(zip(dates, price), columns=['Date', 'Company_A'])

print(f"Returns\n{returns}\n")
print(f"Price\n{price}\n")
print(f"Dates\n{dates}\n")
print(f"Data\n{df}")
```

```python
# Chart

fig = px.line(
    df,
    x='Date',
    y='Company_A',
    title='Company A stock price',
    labels={'Date': 'Date', 'Company_A': 'Price'}
)

fig.show()
```
